## Format Terminal in linux

When opening the terminal you are in the home directory, the command "cd" will take you here, and the command "ls" lists out all the folders, files, etc. in this directory.  The list generated by "ls" is in a default format and color scheme.  Now, the command "ls -F" will generate the list in the format we want, however this is combersome, so we will want to now change the command "ls" to be shorthand for "ls -F" which can be done with the command 'alias ls "/bin/ls -F"'.  Now the command ls has the correct format, but the list is no longer colored, so we need to add more to our "alias" command.  We can again type 'alias ls "/bin/ls -F --color=auto"' and the ls comand is now both formatted correctly, and has the correct color.  However, this is completely local to this terminal tab, if we open a new tab, the it will have the default formatting, to fix this we can write to the ".cshrc" file, which has a set of commands that are carried out when a new terminal tab is opened.  We can use the text editor "emacs" to do this by tryping "emacs .cshrc" which will open the ".cshrc" file where we can save the command 'alias ls "/bin/ls -F --color=auto"' and now every new tab we open will have this formatting.

## Hidden Files and Folders

The command "ls" shows a list of files in the directory you are currently in.  But that is not all, the command "ls -a" will show all the files and folders in that directory.  Hidden files and folders have a "." at the biginning of their names, for example the file that runs through a list of specified commands when a new terminal is opened, ".cshrc" is a hidden file.  "./" is a reference to the current directory, and "../" is a reference to the previous directory, two hidden files that all directories have.

## Common Comands in Terminal

 * "cd" -Change directory.  Will take to home directory without additional specification such as a space and the directory you want to go to
 * "[tab]" -Autocomplete.  Will autocomplete the work you are typing if there is only one possible thing you could be typing your current directory, otherwise it lists out all possible directory names you could be typing
 * "ls" -lists out all visible sub-directories and files contained in the current directory
 * "ls -a" lists out ALL sub-directories and files contained in the current directory, including hidden ones.
 * "mkdir [directory name]" -Creates a sub-directory contained in the current directory by the name of "directory name".

## Updating ipython scripts to github

Once you have a ipython notebook you are content with, it is time to upload or update that notebook to the github server.  To upload the file, simply type the command "git status" which will produce the following output:

    # On branch master
    #
    # Initial commit
    #
    # Untracked files:
    #   (use "git add <file>..." to include in what will be committed)
    #
    #	.ipynb_checkpoints/
    #	[filename].ipynb
    nothing added to commit but untracked files present (use "git add" to track)

This output means that the file "[filename].ipynb" has not been uploaded to the github server before (github is not tracking the file).  Now, we can type the command "git add [filename].ipynb" to allow github to track the file.  Now we can type "git status" to view the updated status:

    # On branch master
    #
    # Initial commit
    #
    # Changes to be committed:
    #   (use "git rm --cached <file>..." to unstage)
    #
    #	new file:   [filename].ipynb
    #
    # Untracked files:
    #   (use "git add <file>..." to include in what will be committed)
    #
    #	.ipynb_checkpoints/

meaning the file has been added, and thus the file is no longer untracked.  However, this is not all, the addition needs to be committed, or simply, the uploader must achknowledge the shown changes with a message.  As shown in the output, this can be done with "git commit [filename].ipynb -m [commit message]" giving the output

    [master (root-commit) <some alphanumeric code>] initial vesion of Introduction ipython notebook
     Committer: [User Name] <user email>

now, if we type the command "git status" we get the output:

    # On branch master
    # Untracked files:
    #   (use "git add <file>..." to include in what will be committed)
    #
    #	.ipynb_checkpoints/
    nothing added to commit but untracked files present (use "git add" to track)

meaning our update were added and have been commit.  Now if we make modifications to the file the command "git status" produces:

    # On branch master
    # Changed but not updated:
    #   (use "git add <file>..." to update what will be committed)
    #   (use "git checkout -- <file>..." to discard changes in working directory)
    #
    #	modified:   Intro.ipynb
    #
    # Untracked files:
    #   (use "git add <file>..." to include in what will be committed)
    #
    #	.ipynb_checkpoints/
    no changes added to commit (use "git add" and/or "git commit -a")

which says explicitly that the file has been modified, and thus we only need to commit the file again (preferable with a new commit message) to update the tracked file.  To find different commit messages along with who commit the file and when they did so, simply type the command "git log".  Now that the file is tracked by github all we need do is push the tracked file onto the github server with the command "git push" however, if this is the first time you are pushing files, it is important to use the command "git push -u origin master" before simply using "git push" so that github knows where to push the file.

### Python

To first connect to the database we simply type

```python
connection = MySQLdb.connect(db = "Kepler", host = "tddb.astro.washington.edu", user = "tddb", passwd = "tddb")
```

which simply making an object that can interact with the data base.  This is done by calling the method connect of the MySQLdb class which connects to the database using the name of the database ("db = Kepler"), specifying the "location" of the database ("host = "tddb.astro.washington.edu"), the user ("user = tddb"), and the password ("passwd = tbbd").  Now we can make a another object that "points to" specific attributes of the connection object with the code

```python
cursor = connection.cursor()
```

Now, we can make a command string that the cursor object will execute when passed as an argument to the execute method.  if the command string is

```python
command = "select * from source where keplerid = 1723618"
```

where "1723618" refers to some star, the execute method will point to all the data points Kepler took for that star.  However this is a lot of data points, we can select specific data, time, flux, and flux error (needed for created light curves) with the string 

```python
command = "select time, sap_flux, sap_flux_err from source where keplerid = 1723618"
```

which tell the cursor to point to the time, flux, and flux error in the database.  There are 11 different types of data for each star, they can be found by typing "describe source;" when accesing the database in terminal (see Terminal section for directions).  Now that our cursor is pointing to these data, we can assign all the values in a variable "results" and then make a list of time data points, sap_flux data points, and sap_flux_err data points by writing.

```python
cursor.execute(command)
results = cursor.fetchall()
time = [result[0] for result in results]
flux = [result[1] for result in results]
dflux = [result[2] for result in results]
```

Now that we have data, we can make a light curve.  We need to use the plotting package of pythond to do so.  This package (matplotlib.pyplot) can be imported similarly to the MySQR package, but with an added trick.  

```python
import matplotlib.pyplot as plt
```

the "as plt" is added to this import so that when we want to use the package we dont need to write out the full name, but only plt.  So, to plot the flux against the time as red circles (with flux error bars around each data point), we type

```python
plt.errorbar(time, flux, yerr = dflux, fmt = "ro")
plt.show()
```

where the last line simply tells the computer to show the plot.  Now, altogether we can plot a light curve as follows:

```python
import MySQLdb
import matplotlib.pyplot as plt
connection = MySQLdb.connect(db = "Kepler", host = "tddb.astro.washington.edu", user = "tddb", passwd = "tddb")
cursor = connection.cursor()
# command = "select * from source where keplerid = 1723618"
command = "select time, sap_flux, sap_flux_err from source where keplerid = 1723618"
cursor.execute(command)
results = cursor.fetchall()
# print results
time = [result[0] for result in results]
flux = [result[1] for result in results]
dflux = [result[2] for result in results]
plt.errorbar(time, flux, yerr = dflux, fmt = "ro")
plt.show()
```
