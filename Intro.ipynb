We must first get the code for the model.  This can be found on github.  We must clone the file on github so that is it executable on your current computer. This command is "f2py -c occultquad.f -m occultquad" which creates the compiled code file, "occultquad.so*". *Now, we can import this file into our python code so that we can use it, but wait, simply importing "occultquad" does not work. You can also edit the environments $PYTHONPATH in ".cshrc" as we did to format out terminal, but this work fine for our purposes.  Now, we may created our own light curves using the model.  You will notice in the current directory a file named "tqfortran.py" which is a definition of a wrapper class that allows us to use the model in python.  We can simply copy its contents into our python script as follows.

```python
import numpy as np

def ModLC(timeIn, RpRs, b, tR, t0, u1, u2):
    bsq  = b**2
    z0   = np.sqrt(bsq + ((timeIn-t0) / tR)**2)
    nz   = len(timeIn)
    muo1 = np.zeros((nz))
    mu0  = np.zeros((nz))
    _    = occultquad.occultquad(z0,u1,u2,RpRs,muo1,mu0,nz)
    return muo1
```

Now, we are ready to create our own model light curves.  The model has 5 input parameters; RpRs, b, tR, t0, u1, and u2.  Here, timeIn is simply the time we want to create the curve over.  RpRs is the ratio of the planets radius to the radius of the Star, b is the minimum impact parameter, meaning that as the planet transits the star it does not necessarily have to go directly across the center of the star (b = 0), but rater it can go transit at any distance from the center.  This distance, the center of the planet to the center of the star is the systems impact parameter,  t0 is the time at which the planet achieves its minimum impact parameter b, tR is the time it takes the planet to transit one radius of the star, and finally, u1/u2 are the parameters that describe the stars Limb darkening.  Limb darkening is simply the fact that stars are not of uniform brightness throughout, and in fact tend to be less bright toward the edges.  For more information on these parameters see the paper "Analytic Light Curves for Planetary Transit Searches" by Mandel and Agol.  We must decide what are timeIn will be and choose values for all other parameters, then we may call our method ModLC defined in "tqfortran.py" to create the brightness data points which are defined to be 1 when there is no transit occurring, and thus we have the light curve 

```python
timeIn = np.arange(-1, 2, 0.1)
Model = ModLC(timeIn, 0.01, 0, 0.5, 0, 0.1, 0.1)

import matplotlib.pyplot as plt
plt.plot(timeIn, Model, 'ro')
plt.show()
```

Now by simply varying the parameters we may discover how each affects the light curve, first lets vary RpRs and b.

```python
RpRs = [0.001, 0.01, 0.1, 1]
b = [0, 0.25, .5, 0.75]

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row', figsize = (12,8))
for x in range(0,4):
    for y in range(0,4):
        Model = ModLC(timeIn, RpRs[x], b[y], 0.5, 0, 0.1, 0.1)
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('RpRs=' + repr(RpRs[x]) + ",b =" + repr(b[y]))

plt.show()
```

So, since in this plot we varied the radius and the impact parameter, it is evident that a larger impact parameter decreases the length of transit, which makes sense because if the planet transits at a higher "latitude" there will be less distance to cover.  We also see that increasing the impact parameter decreases the depth of the light curve ever so gradually.  Finally, we see that increasing the radius of the planet with respect to the radius of the star also decreases the depth of the light curve.  Next lets vary tR and t0. 

```python
tR = [0.25, 0.5, 0.75, 1]
t0 = [-0.5, 0, 0.5, 1]

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row', figsize = (12,8))
for x in range(0,4):
    for y in range(0,4):
        Model = ModLC(timeIn, 0.01, 0, tR[x], t0[y], 0.1, 0.1)
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('tR=' + repr(tR[x]) + ",t0 =" + repr(t0[y]))

plt.show()
```

Again, it is evident that by increasing tR the length of the transit increases which makes since because it takes longer to travel the same distance, and if we vary t0, the light curves position changes with respect to the other data points.  Now, for the interesting one, the Limb darkening.

```python
u1 = [0, 0.25, 0.5, 0.75]

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row', figsize = (12,8))
for x in range(0,4):
    for y, u2 in enumerate ([-u1[x], 0.25 - u1[x], 0.75 - u1[x], 1 - u1[x]]): 
        Model = ModLC(timeIn, 0.01, 0, 0.5, 0, u1[x], u2)
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('u1=' + repr(u1[x]) + ",u2 =" + repr(u2))

plt.show()
```

Now, as we can see when u1 = u2 = 0 and there is no Limb darkening, after the planet is completely in front of the interior of the star, the curve is flat, and thus no changes in brightness, and that as u1 and u2 increase, so does how curved the base of the curve is.  Now, it is important to know that both u1 and u2 have constraints and assigning values that violate these constraints lead to odd curve characteristics such as limb brightening.  These constraints are as follows: 0 < u1, and 0 < u1 + u2 < 1.

## Fitting 

Notice, all of these graphs have normalized flux, and thus to fit our data to an actual light curve we must scale the model by multiplying by the unobstructed flux of the star, A.  Now that we know how each parameter of the model change the shape of the light curve, we are able to begin fitting out actual data to the model in order to get an idea for what parameters actually characterize that system.  To do this we must first convert our list of data to arrays to simply make them easier to work with.

```python
time_one = np.array(time_one)
flux_one = np.array(flux_one)
dflux_one = np.array(dflux_one)
```

Now that we have our data in the form of arrays we can create a function that gives us a sense of how good a fit those specific parameters are.  For our sake this value will be defined as $$\chi^{2} = \sum_{data} \bigg( \frac{Model - data} {uncertainty} \bigg) ^2$$  However, there are some caveats with this function; there are limits on some of the parameters, but this does not take into account what happens when these limits are violated so in our function definition we must include a conditional statement that returns an absurdly large value to avoid a fit to the data with those incorrect parameters.  To begin, we give the function a name and specify the inputs of the function, in this case an array of parameters of our model (it is important to remember the order), and an array of arguments (the data in our case).  Now, we must set the specific parameters to their corresponding variables so we can create a model, set up the conditional, and assign variables to the data.  Now that we have everything correctly define we need simply set of the scaled model with specific parameters, calculate the $\chi^2$ value, and return it.

```python
def chi2(params, *args): 
    A, RpRs, b, tr, t0, u1, u2 = params
    if RpRs < 0 or RpRs > 1 or b < 0 or b > 1 or tr < 0 or u1 < 0 or (u1 + u2) < 0 or (u1 + u2) > 1:
        return 1e10
    x, y, dy = args
    Model = A * ModLC(x, RpRs, b, tr, t0, u1, u2)
    residual = Model - y
    sigma = residual / dy
    return np.sum(sigma**2)
```

Now that our $\chi^2$ function is set up, we need a way to minimize this value, which can often be a very difficult process.  Luckily there are minimizing functions already created that we must simply import.  This can be done with the line "from scipy.optimize import fmin, fmin_bfgs" prior to fitting any data. Next we simply need set up our initial guess for the fit and assign the results of the fit to a variable and print it.  It is important that our initial guess be somewhat accurate so that the minimizing has some idea of where to go, a bad guess can cause the minimizing algorithm to converge to a bad fit, not converge at all, or simply time out.  In this example we are using the data that we previously imported, so we can get a general idea for the important parameters from analyzing the graph of its light curve.

```python
from scipy.optimize import fmin, fmin_bfgs


guess = np.array([1786500, 0.01, 0.5, 0.05, 122.3, 0.5, 0.25])
results = fmin_bfgs(chi2, x0 = guess, args = (time_one, flux_one, dflux_one))
print results
```

As it is evident, the model parameters did change.  An easy way to qualitatively evaluate the fit is simply to plot it against the data, including the initial guess to see how much the minimizing algorithm changed the curve. 

```python
plt.errorbar(time_one, flux_one, dflux_one, fmt = "ro")
model = guess[0] * ModLC(time_one, guess[1], guess[2], guess[3], guess[4], guess[5], guess[6])
plt.plot(time_one, model, "k-")

model = results[0] * ModLC(time_one, results[1], results[2], results[3], results[4], results[5], results[6])
plt.plot(time_one, model, "b-")

print results 
```

Here the data is red dots, the fit is a blue line, and the initial guess is the black line.  Clearly this is a better fit than the initial guess, but we can do better.  Our time of transit looks a little off from our curve, so we can check if the initial guess was close to the actual time.  Remember our initial guess was 122.3.

```python
print time_one
```

Sure enough, our initial guess does fall within the bounds of the data, so this could not have contributed that greatly to the badness of the fit.  One interesting thing that could affect the goodness of fit is the limb darkening variable.  They have a fairly small window of acceptable values, and the wrong values affect the shape of the curve dramatically.  Now, they are not very well defined by the data since there are so few data points inside the actual transit, plus they are fairly similar for many system, so what happens if we hold these values constant at our initial guess?  To do this we must slightly change our $\chi^2$ function to account for the constant u1 and u2.

```python
u1_fix = 0.5
u2_fix = 0.25

def chi2_nold(params, *args): 
    A, RpRs, b, tr, t0 = params
    if RpRs < 0 or RpRs > 1 or b < 0 or b > 1 or tr < 0:
        return 1e10
    x, y, dy = args
    Model = A * ModLC(x, RpRs, b, tr, t0, u1_fix, u2_fix)
    residual = Model - y
    sigma = residual / dy
    return np.sum(sigma**2)
```

Now, we can simply re-fit the data to the model in the same way we did earlier, but without specifying u1 and u2.  Since there are 2 fewer parameters, we should expect a lower $\chi^2$ thus a simple comparison of the two numbers is not an effective measure of the goodness of the fit.  Thus, we must simply plot the fit to examine qualitatively.

```python
guess = np.array([1786500, 0.01, 0.5, 0.05, 122.3])
results = fmin_bfgs(chi2_nold, x0 = guess, args = (time_one, flux_one, dflux_one))

plt.errorbar(time_one, flux_one, dflux_one, fmt = "ro")
model = guess[0] * ModLC(time_one, guess[1], guess[2], guess[3], guess[4], u1_fix, u2_fix)
plt.plot(time_one, model, "k-")

model = results[0] * ModLC(time_one, results[1], results[2], results[3], results[4], u1_fix, u2_fix)
plt.plot(time_one, model, "b-")

print results 
```

This is obviously a better fit, and thus we can conclude that in the past attempt to fit the data, the minimizing algorithm got stuck varying the values of u1 and u2.

Unlike our previous method for fitting that data which minimized the value of $\chi^2$, this new method uses a function $$\ln(likelihood) = -0.5 \chi^2$$  Which will eventually allow us to get a better estimate on the uncertainty of each parameter value.  We can define this function as follows

```python
def lnlike(params, *args):
    return -0.5 * chi2_nold(params, *args)
```

Now that we have the $\ln(likelihood)$ function, we must again initialize our guess as well as some other values.  For this example we can use the results of our previous fit as our initial guess.  We need to now specify the dimensions of our parameter space, or simply the number of parameters that need to be fitted to our model.  We also need to specify the number of walkers.  It is a good rule of thumb to have far more walkers than parameters.  The reason they are called walkers is that in a sense "walk" around the parameter space, taking far more steps at places with high values of $\ln(likelihood)$ and very few steps at locations with low $\ln(likelihood)$ value.  Finally, we must initialize the walkers position, as well as specify the number of initial steps to throw away that could skew the fit, and how many steps we want them to take in total.

```python
guess = results
ndim = len(guess)
nwalkers = 2 * ndim
nburn = 100
nstep = 10000

pos = [guess + 1e-3 * np.random.randn(ndim) for i in range(nwalkers)]

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlike, args = (time_one, flux_one, dflux_one))
pos, prob, state = sampler.run_mcmc(pos, nburn)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, nstep, rstate0 = state)
```
