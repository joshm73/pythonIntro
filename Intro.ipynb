## Format Terminal in linux

When opening the terminal you are in the home directory, the command "cd" will take you here, and the command "ls" lists out all the folders, files, etc. in this directory.  The list generated by "ls" is in a default format and color scheme.  Now, the command "ls -F" will generate the list in the format we want, however this is combersome, so we will want to now change the command "ls" to be shorthand for "ls -F" which can be done with the command 'alias ls "/bin/ls -F"'.  Now the command ls has the correct format, but the list is no longer colored, so we need to add more to our "alias" command.  We can again type 'alias ls "/bin/ls -F --color=auto"' and the ls comand is now both formatted correctly, and has the correct color.  However, this is completely local to this terminal tab, if we open a new tab, the it will have the default formatting, to fix this we can write to the ".cshrc" file, which has a set of commands that are carried out when a new terminal tab is opened.  We can use the text editor "emacs" to do this by tryping "emacs .cshrc" which will open the ".cshrc" file where we can save the command 'alias ls "/bin/ls -F --color=auto"' and now every new tab we open will have this formatting.

## Hidden Files and Folders

The command "ls" shows a list of files in the directory you are currently in.  But that is not all, the command "ls -a" will show all the files and folders in that directory.  Hidden files and folders have a "." at the biginning of their names, for example the file that runs through a list of specified commands when a new terminal is opened, ".cshrc" is a hidden file.  "./" is a reference to the current directory, and "../" is a reference to the previous directory, two hidden files that all directories have.

## Common Comands in Terminal

 * "cd" -Change directory.  Will take to home directory without additional specification such as a space and the directory you want to go to
 * "[tab]" -Autocomplete.  Will autocomplete the work you are typing if there is only one possible thing you could be typing your current directory, otherwise it lists out all possible directory names you could be typing
 * "ls" -lists out all visible sub-directories and files contained in the current directory
 * "ls -a" lists out ALL sub-directories and files contained in the current directory, including hidden ones.
 * "mkdir [directory name]" -Creates a sub-directory contained in the current directory by the name of "directory name".
