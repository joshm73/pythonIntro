## Planetary Transit Model

We must first get the code for the model.  This can be found on github.  We must clone the file on github so that is it executable on your current computer.  This can be done with the command "git clone [address]" and in the case, the address of the repository we need is "git@github.com:acbecker/EXOQ.git". Now, the "tail [file]" command will run the file and print out the last few lines of the code.  It is possible to specify the number of lines by using "tail -x [file]" in which case, the last x lines are printed.  If we "tail -2 occultquad.f" the final line of output is a string that gives the command for converting the fortran code into something usefull for us in python, comiled bionary commands that the computer can read, and is in fact faster at reading.  This command is "f2py -c occultquad.f -m occultquad" which creates the compiled code file, "occultquad.so*". *Now, we can import this file into our python code so that we can use it, but wait, simply importing "occultquad" does not work. You can also edit the environments $PYTHONPATH in ".cshrc" as we did to format out terminal, but this work fine for our purposes.  Now, we may created our own light curves using the model.  You will notice in the current directory a file named "tqfortran.py" which is a definiton of a wrapper class that allows us to use the model in python.  We can simply copy its conetents into our python script as follows.

```python
import numpy as np

def ModLC(timeIn, RpRs, b, tR, t0, u1, u2):
    bsq  = b**2
    z0   = np.sqrt(bsq + ((timeIn-t0) / tR)**2)
    nz   = len(timeIn)
    muo1 = np.zeros((nz))
    mu0  = np.zeros((nz))
    _    = occultquad.occultquad(z0,u1,u2,RpRs,muo1,mu0,nz)
    return muo1
```

Now, we are ready to create our own model light curves.  The model has 5 input parameters; RpRs, b, tR, t0, u1, and u2.  Here, timeIn is simply the time we want to create the curve over.  RpRs is the ratio of the planets radius to the radius of the Star, b is the minimum impact parameter, meaning that as the planet transits the star it does not neccesarily have to go directly across the center of the star (b = 0), but rater it can go transit at any distance from the center.  This distance, the center of the planet to the center of the star is the systems impact parameter,  t0 is the time at which the planet achieves its minimum impact parameter b, tR is the time it takes the planit to transit one radius of the star, and finally, u1/u2 are the parameters that describe the stars Limb darkening.  Limb darkening is simply the fact that stars are not of uniform brightness throughout, and in fact tend to be less bright toward the edges.  For more information on these parameters see the paper "Analytic Light Curves for Planetary Transit Searches" by Mandel and Agol.  We must decide what are timeIn will be and choose values for all other parameters, then we may call our method ModLC defined in "tqfortran.py" to create the brightness data points which are defined to be 1 when there is no transit occuring, and thus we have the light curve 

```python
timeIn = np.arange(-1, 2, 0.1)
Model = ModLC(timeIn, 0.01, 0, 0.5, 0, 0.1, 0.1)

import matplotlib.pyplot as plt
plt.plot(timeIn, Model, 'ro')
plt.show()
```

Now by simply varying the parameters we may discover how each affects the light curve, first lets vary RpRs and b.

```python
RpRs = [0.001, 0.01, 0.1, 1]
b = [0, 0.25, .5, 0.75]

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row')
for x in range(0,4):
    for y in range(0,4):
        Model = ModLC(timeIn, RpRs[x], b[y], 0.5, 0, 0.1, 0.1)
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('RpRs=' + repr(RpRs[x]) + ",b =" + repr(b[y]))

plt.show()
```

So, since in this plot we varried the radius and the impact parameter, it is evident that a larger impact parameter decreases the length of transit, which makes sense because if the planet transits at a higher "latitude" there will be less distance to cover.  We also see that increasing the impact parameter decreases the depth of the light curve ever so gradually.  Finally, we see that increasing the radius of the planet with respecto to the radius of the star also decreases the depth of the light curve.  Next lets vary tR and t0. 

```python
tR = [0.25, 0.5, 0.75, 1]
t0 = [-0.5, 0, 0.5, 1]

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row')
for x in range(0,4):
    for y in range(0,4):
        Model = ModLC(timeIn, 0.01, 0, tR[x], t0[y], 0.1, 0.1)
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('tR=' + repr(tR[x]) + ",t0 =" + repr(t0[y]))

plt.show()
```

Again, it is evident that by increasing tR the length of the transit increases which makes since because it takes longer to travel the same distance, and if we vary t0, the light curves position changes with respect to the other data points.  Now, for the interesting one, the Limb darkening.

```python
u1 = [0, 0.25, 0.5, 0.75]
u2 = u1

f, axarr = plt.subplots(4, 4, sharex='col', sharey='row')
for x in range(0,4):
    for y in range(0,4):
        Model = ModLC(timeIn, 0.01, 0, 0.5, 0, u1[x], u2[y])
        axarr[x, y].plot(timeIn, Model, 'ro')
        axarr[x, y].set_title('u1=' + repr(u1[x]) + ",u2 =" + repr(u2[y]))

plt.show()
```

Now, as we can see when u1 = u2 = 0 and there is no Limb darkening, after the planet is completely in front of the interior of the star, the curve is flat, and thus no changes in brightness, and that as u1 and u2 increase, so does how curved the base of the curve is.  Finally, at high values of u1 and u2, the light curve increases meaning the star gets brighter when the planet is just passing over the edge of the star.
